```python
def get_multiplication(argument):
    x_list = argument[0]
    y_list = argument[1]
    result = []
    
    if not x_list or not y_list:
        return result
    
    for x in x_list:
        for y in y_list:
            result.append(round(x * y/100, 2))
    return result
```

```python
audience_df['scores_per_age_genders'] = audience_df[['scores_per_age', 'num_per_genders']].apply(get_multiplication, axis=1)
audience_df
```

```python
audience_df['num_per_ages_genders'] = audience_df[['num_per_ages', 'num_per_genders']].apply(get_multiplication, axis=1)
audience_df['scores_per_ages'] = audience_df[['scores_per_age', 'num_per_ages']].apply(get_multiplication, axis=1)
audience_df
```

#### scores_per_age_genders
scores_per_age * num_per_genders : 나이대별 점수 * 성별 인원 비율

ex) 10대 점수 * 남성 비율, 20대 점수 * 남성 비율, ..., 40대 점수 * 여성 비율, 50대 이상 점수 * 여성 비율 => 10개

#### scores_per_ages
scores_per_age * num_per_ages : 나이대별 점수 * 나이별 인원 비율

ex) 10대 점수 * 10대 비율, 20대 점수 * 20대 비율, ..., 50대 이상 점수 * 50대 이상 비율 => 5개

#### num_per_ages_genders
num_per_genders * num_per_ages : 성별 인원 비율 * 나이별 인원 비율

ex) 남자 비율 * 10대 비율, 여자 비율 * 10대 비율, 남자 비율 * 20대 비율, 여자 비율 * 20대 비율, ..., 여자 비율 * 50대 이상 비율 => 10개

```python
def get_row(arg):
    result = [arg[0]]
    if arg[1]:
        result.extend(arg[1])
    else:
        result.extend([np.nan] * 10)
    return result
```

```python
spag_columns = ['code']
for age in ages:
    for gender in genders:
        spag_columns.append(f'spa_{age}s_npg_{gender}')

spag_df = pd.DataFrame(columns=spag_columns)

for s in audience_df[['code', 'scores_per_age_genders']].apply(get_row, axis=1):
    try:
        spag_df = spag_df.append(pd.Series(s, index=spag_columns), ignore_index=True)
    except:
        unexpected = [s[0]].extend([np.nan] * 10)
        spag_df = spag_df.append(pd.Series(unexpected, index=spag_columns), ignore_index=True)

spag_df
```

```python
npag_columns = ['code']
for age in ages:
    for gender in genders:
        npag_columns.append(f'npa_{age}s_npg_{gender}')

npag_df = pd.DataFrame(columns=npag_columns)

for s in audience_df[['code', 'num_per_ages_genders']].apply(get_row, axis=1):
    try:
        npag_df = npag_df.append(pd.Series(s, index=npag_columns), ignore_index=True)
    except:
        unexpected = [s[0]].extend([np.nan] * 10)
        npag_df = npag_df.append(pd.Series(unexpected, index=npag_columns), ignore_index=True)

npag_df
```

```python
spa_columns = ['code']
for age1 in ages:
    for age2 in ages:
        spa_columns.append(f'spa_{age1}s_npa_{age2}s')

spa_df = pd.DataFrame(columns=spa_columns)

for s in audience_df[['code', 'scores_per_ages']].apply(get_row, axis=1):
    try:
        spa_df = spa_df.append(pd.Series(s, index=spa_columns), ignore_index=True)
    except:
        unexpected = [s[0]].extend([np.nan] * 10)
        spa_df = spa_df.append(pd.Series(unexpected, index=spa_columns), ignore_index=True)

spa_df
```

```python
spa_df = spa_df[['code', 'spa_10s_npa_10s', 'spa_20s_npa_20s', 'spa_30s_npa_30s', 'spa_40s_npa_40s', 'spa_50s_npa_50s']]
```
